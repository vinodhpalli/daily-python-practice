# Inheritence

## single inheritence

```python
class father:
    def driving(self):
        print("driving")
class son(father):
    def study(self):
        print("study")
s=son()
s.driving()
s.study()
        
```

## multiple inheritence

```python
class father:
    def driving(self):
        print("driving")
class mother:
    def cooking(self):
        print("cooking")
class child(father,mother):
    def study(self):
        print("study")
s=child()
s.driving()
s.cooking()
s.study()
               
```

# multilevel 

```python
class Grandfather:
    def farming(self):
        print("farming")
class father(Grandfather):
    def driving(self):
        print("driving")
class child(father):
    def study(self):
        print("study")
s=child()
s.driving()
s.farming()
s.study()
```

## hirarichal inheritence

```python
class father:
    def family_name(self):
        print("family name")
class son(father):
    def study(self):
        print("study")
class daughter(father):
    def dance(self):
        print("dance")       
s=son()
s.family_name()
s.study()
d=daughter()
d.family_name()
d.dance()
```

# abstraction

```python
# in this we have both abstraction and polymorphism
from abc import ABC,abstractmethod
class vechicle(ABC):
    @abstractmethod
    def start(self):
        pass 
    def stop(self):
        print("car stopped")
#polymorphism
class bike(vechicle):
    def start(self):
        print("bike started")
class car(vechicle):
    def start(self):
        print("car started")
c=car()
c.start()
c.stop()
b=bike()
b.start()
b.stop()
```

# Types of polymorphism

## method overloading

```python
#method overloading
class A:
    def add(self,a,b):
        return a+b
    def add(self,a,b,c):
        return a+b+c
    def add(self,a,b,c,d):
        return a+b+c+d
b=A()
print(b.add(2,3,4,5))
```

```python
class A:
    def add(self,*args):
        return sum(args)
b=A
print(b.add(1,2,3,4,5,6))
```

# constuctor overloading

```python
#same constructor--->
class person:
    def __init__(self,name,age=None):
        self.name=name
        self.age=age
p1=person("john")
print(p1.name,p1.age)
p2=person("john",25)
print(p2.name,p2.age)
```

```python
#method overloading--->same method name,different parameters
#operator overloading---->same operator behave differently for diffent data types
#constuctor overloading--->same class constuctor works differently depending on arguments
```

# Run time polymorphism

### Dynamically typed(duck typing)

```python
class dog:
    def speak(self):
        print("bark")
class cat:
    def speak(self):
        print("meow")
class duck:
    def speak(self):
        print("quack") 
def make_sound(animal):
    animal.speak()
    
d=dog()
c=cat()
k=duck()
make_sound(d)
make_sound(c)
make_sound(k)
```

## method over riding

```python
class animal:
    def speak(self):
        print("some sound")
class dog(animal):
    def speak(self):
        print("bark")
d=dog()
d.speak()
    
```

```python
class animal:
    def speak(self):
        print("some sound")
class dog(animal):
    def speak(self):
        super().speak() # cll parent method
        print("bark")
d=dog()
d.speak()
```

## constructor overriding

```python
class parent:
    def __init__(self):
        print("parent constructor")
class child(parent):
    def __init__(self):
        super().__init__() # call parent counstroctor
        print("child constructor")
obj=child()
```

## inheritence--->acessing varibles and methods

```python
class parent:
    def __init__(self):
        self.name="parent class"
        self.value=100
    def show(self):
        print("this is parent clss methosd")
        print("name:",self.name)
        print("value:",self.value)
class child(parent):
    def __init__(self):
        super().__init__() # call parent counstroctor
        self.child_name="child class"
    def display(self):
        print("this is parent clss methosd")
        print("name:",self.child_name)
obj=child()
obj.show()
```

## getter and setter method

```python
class student:
    def __init__(self,name,marks):
        self.__name=name
        self.__marks=marks
    #public method(getter)
    def get_marks(self):
        return self.__marks
    def set_marks(self,new_marks):
        if 0<= new_marks <=100:
            self.__marks=new_marks
        else:
            print("Invalid marks!")
s1=student("john",82)
print(s1.get_marks())
s1.set_marks(95)
print(s1.get_marks())
```

# Encapsulation

```python
class bank:
    def __init__(self,owner,balance):
        self.__owner=owner
        self.__balance=balance
    def get_owner(self):
        return self.__owner
    def get_balance(self):
        return self.__balance
    def deposit(self,amount):
        if amount>0:
            self.__balance+=amount
            print(f"Deposited {amount},New Balance={self.__balance}")
        else:
            print("deposited amount must be positive")
    def withdraw(self,amount):
        if amount>0<=self.__balance:
            self.__balance-=amount
            print(f"withdraw {amount},New Balance={self.__balance}")
        else:
            print("invalid withdraw")
account=bank("john",1000)
print(account.get_owner())
print(account.get_balance())
account.deposit(200)
print(account.get_balance())
account.withdraw(200)
print(account.get_balance())
```
