```python
#Write a program to reverse a number using a while loop.
n =int(input("Enter the number:"))
rev = 0
while n > 0:
    digit = n % 10
    rev = rev *10 + digit
    n //= 10
print(rev)
```

```python
#Write a program to check whether a number is palindrome.
n = int(input("Enter the string:"))
rev = 0
while n >0:
    digit = n % 10
    rev = rev*10 + digit
    n //= 10
    if n == rev:
        print("palindrome")
print("Not palindrome")
```

### Write a program to find the largest digit in a number.

```python
n = int(input("Enter the number:"))
largest = 0
while n>0:
    digit = n % 10
    if digit > largest:
        largest = digit
    n = n//10
print("largest element is :" , largest)
```

#### Write a program to count uppercase and lowercase letters in a string.

```python
a = input("Enter the stirng")
count = 0
count_1 = 0
for i in a:
    if i.islower():
        count+=1
    elif i.isupper():
        count_1+=1
    elif i.isdigit():
        print(f" {i} is an digit")
print("Total lower case" , count)
print("Total upper case" , count_1)
```
