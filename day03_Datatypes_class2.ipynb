# String Methods

```python
text  =  "Hello World!"
```

```python
# len()--> Returns the length of the string
print("Length of string:", len(text))
```

```python
# lower() ---> Converts the string to lowercase
print("Lowercase:",text.lower())
```

```python
# upper() ---> Converts the string to uppercase
print("Uppercase:",text.upper())
```

```python
# srtip() ---> Removes leading and trailing spaces 
text = "       Hello,World!"
text1 = "Hello,World!    "
print(text)
print(text1.strip())
print(text.strip())
```

```python
# Replace() ---> Replaces a word or chareacter with another
text  =  "Hello World!"
print("Text:",text)
print("Replace world with python:",text.replace("World" , "python"))
print("--------------------------")
text1 = "Hello"
print(text1.replace("e",'a'))
print(text1.replace("l" , "j"))
```

```python
#split() -----> Splits the string into list based on a separator
text = 'apple,banana,mango'
print(text.split(","))
a = 'apple','banana','mango'
print(a)
```

```python
# startswith() ---> checks if string starts with a given prefix
text = "HelloWold"
print(text.startswith("He"))
print(text.startswith("je"))
```

```python
# endswith() ---> checks if string ends with a given suffix
text  = "HelloWorld"
print(text.endswith("d!"))
print(text.endswith("d"))
```

```python
# find() ---> Returns the index of the first occurence of a substring
text = "HelloWolrd!"
print(text.find('o'))
```

```python
# count() ---> Counts how many times a substring occurs
text = "Hello,Hello,World"
print(text.count("Hello"))
```

```python
# isaplha() ---> cehcks if all characters are letters
text = "Hello"
print(text.isalpha())
text1 = "Hello1"
print(text1.isalpha())
```

```python
#isdigit() ---> checks if all the charecters are digits
text = '123'
print(text.isdigit())
text1  = "123he"
print(text1.isdigit())
```

```python
#isalnum() ---> checks if the charecters are letters or numbers
text = "Hello123"
print(text.isalnum())
```

```python
# capitalize--> Give first charecters in upper case
str.capitalize("hello world")
```

## Format Function

```python
name = 'john'
age = 20
print("My name is {} and I am {} years old".format(name,age)) # used to print bot values in single line
```

```python
s = "My name is  {a} and I am {b} years old".format(a=name,b=age)
print(s)
```

```python
print("{} and {} are best friends".format("virat","rohit"))
```

```python
print("{2},{0} and {1} are best friends".format("Virat","Rohit","Dravid"))
```

```python
# %d ----> Takes all the digits without decimals
# %f ----> Takes all the decimals digits 
# %s ----> Takes all the stirng elements
Integer = 10
Float = 10.5
String = 'Hari'
print("Hi I am an Integer...My value is %d\nHi I am Float....My value is %f\nHi I am an String....My value is %s"%(Integer,Float,String))
```

## Indexing

```python
a = [10,20,30,40,50]
print(type(a))
print(a[0])
print(a[1])
print(a[2])
print(a[3])
print(a[4])
```

```python
a = [10,20,30,40,50]
print(type(a))
print(a[-1])
print(a[-2])
print(a[-3])
print(a[-4])
print(a[-5])
```

```python
# immutable ---> cannot change elements
a = "Hello"
print(a[1])
a[1]='a'
```

```python
#list is mutable
a = [1,2,3,4,5]
print(a[1])
a[1]="Hello"
a[4] = "hi"
print(a)
```

```python
a = [10,20,30,40,50]
print(a[0:4])
print(a[::-1])
print(a[0:5:2])
print(a[-5:])
```

```python
print(a)
print(a[1:3])
a[1:3]=[85,86]
print(a)
```

```python
a = [10,20,30,40,50]
print(a[-2])
a[-2]="hello"
print(a)
```
