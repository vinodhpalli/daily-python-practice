```python
# multipe if
age = int(input("Enter the age:"))
if age <18:
    print("You are not eligible for vote")
if age >=18:
    print("You are eligible for vote")
if age >= 21:
    print("You are eligible for Marriage")
if age >= 60:
    print("You are eligible senior citizen benifits")
if age >= 80:
    print("You are very senior citizen, Take Extra care of your health")
```

```python
# check if a number is Even or not

number = int(input("Enter the number:"))
if number % 2 == 0:
    print("Even Number")
```

```python
# Find the greatest of three numbers

a = int(input("Enter a 'a' value:"))
b = int(input("Enter a  'b' value:"))
c = int(input("Enter a 'c' value:"))
if a>b and a>c:
    print("a is greater")
if b>a and b>c:
    print("c is greater")
if c>a and c>b:
    print("c is greater")
        
```

# if - else Condition

```python
age = 16

if age >=18:
    print("You are eligible for vote")
else:
    print("You are not eligible for vote")
```

```python
# even or odd checker

number  = int(input("Enter a numbers:"))
if number % 2 == 0:
    print(f"{number} is an even number")
else:
    print(f"{number} is an odd number")
```

```python
# pass or fail checker
marks  = int(input("Enter the marks:"))

if marks >= 35:
    print("You have passed")
else:
    print("You have failed")
```

# if-elif-else statement

```python
temperature  = int(input("Enter the temperature:"))
if temperature > 30:
    print("It is hot day")
elif temperature > 20:
    print("It is warm day")
elif temperature > 10:
    print("It is cool day")
else:
    print("It is cold day")
            
```

```python
marks = int(input("Enter your marks"))
if marks >= 90:
    print("Grade A")
elif marks >= 80:
    print("Grade B")
elif marks >= 70:
    print("Grade C")
elif marks >= 60:
    print("Grade D")
else:
    print("Grade F")
```
