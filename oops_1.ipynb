```python
class person:
    def __init__(self,name,age):#attributes
        self.name=name
        self.age=age
    def greet(self):#methods
        print(f"Hello,my name is {self.name} and Iam {self.age} years old.")
# create object
person1=person("Amit",22)
person2=person("Vinodh",22)

print(f"{person1.name} is {person1.age} years old")
print(f"{person2.name} is {person2.age} years old")

person1.greet() # calling method
person2.greet()
```

```python
class car:
    def __init__(self,brand,colour,fuel):
        self.brand=brand
        self.colour=colour
        self.fuel=fuel
    def start(self):
        print(f"The {self.colour} {self.brand} with {self.fuel} fuel has started")
    def stop(self):
        print(f"The {self.colour} {self.brand} with {self.fuel} fuel has stoped")
car1=car("Honda","Red","Petrol")
car2=car("Suzuki","Blue","Disel")
print(car1.brand,car1.colour,car1.fuel)
print(car2.brand,car2.colour,car2.fuel)
car1.start()
car1.stop()
car2.start()
car2.stop()
```

# types of methods 

## instance method

```python
class student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def show_details(self):
        print(f"name:{self.name},marks:{self.marks}")
s1=student("Amit",22)
s2=student("Vinodh",100)
s1.show_details()
s2.show_details()
```

```python
#decerator
#class method:modify or acess data common to all objects
class student:
    school_name="abc high school" #class varible
    def __init__(self,name):
        self.name=name
    @classmethod    
    def change_school(cls,new_name):
        cls.school_name=new_name #change all the students
student.change_school("XYZ high school")
s1=student("Amit")
print(s1.school_name)
```

```python
#static method
class student:
    school_name="abc high school" #class varible
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    @staticmethod 
    def caluclate_grade(marks):
        if marks>=90:
            return "A"
        elif marks>=75:
            return "B"  
        else :
            return "C"
s1=student("Vinodh",100) 
s2=student("Amit",75)
print(s1.caluclate_grade(1001))
print(s2.caluclate_grade(75))
```

# access modifiers

## public access modifiers

```python
class student:
    def __init__(self,name,marks):
        self.name=name #public varible 
        self.marks=marks
    def display_info(self):
        print(f"name:{self.name},marks:{self.marks}")
s1=student("Amit",22)
s2=student("Vinodh",100)
print(s1.name,s1.marks)
s1.display_info()
```

# private access modifiers

```python
class student:
    def __init__(self,name,marks):
        self.__name=name #private varible 
        self.__marks=marks #private varible 
    def __caluclate_result(self):#private method 
         if self.__marks>=90:
            return "A"
         elif self.__marks>=75:
            return "B"  
         else :
            return "C"
    def display_result(self):
        print(f"name:{self.__name},marks:{self.__marks}")
        print("result:", self.__caluclate_result())
s1=student("Amit",67)
s2=student("Vinodh",100)
s1.display_result()
s2.display_result()
```

# protacted acess modifier

```python
class student:
    def __init__(self,name,marks):
        self._name=name #protacted varible 
        self._marks=marks #protacted
    def _display_info(self):
        print(f"name:{self._name},marks:{self._marks}")
s1=student("Amit",90)
print(s1._name)#--> still works ,but should be avoid
#s1._display_info()#--->still works ,but should be avoid
```

# INHERITENCE

```python
class animal:
    def eat(self):
        print("eat")
    def sleep(self):
        print("sleep") 
#child class 1
class dog(animal):#dog inherits from the animal
    def bark(self):
        print("Barks")
#child class2
class cat(animal):#cat inherits from the animal
    def meow(self):
        print("Meow")
Dog=dog()
Dog.eat()
Dog.sleep()
Dog.bark()
Cat=cat()
Cat.eat()
Cat.sleep()
Cat.meow()
```
