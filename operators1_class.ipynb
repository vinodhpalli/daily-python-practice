# 1.Arithmetic Operator

```python
a =10
b = 4

print("Addition:" ,a+b)
print("Subtraction:", a-b)
print("Multiplication:" ,a*b)
print("Division:" , a/b)
print("Floor division:" , a//b)
print("Modulus:" , a%b)
print("Exponent:" ,a ** b)
```

## 2.Comparison Operator

```python
a = 5
b = 3
print("a==b",a==b)
print("a!=b",a!=b)
print("a>b" , a>b)
print("a<b" , a<b)
print("a>=b" , a>=b)
print("a<=b" , a<=b)
```

## 3.Assignment Operator

```python
a = 10
b = 4
a+=b
print(a)
a-=b
print(a)
a*=b
print(a)
a//=b
print(a)
a%=3
print(a)
a**=20
print(a)
```

## 4.Logical Operators

```python
a = 7
print(a>5 and a<6)   # true and false--> false
print(a>5 and a!=6)  # true and true ---> true

print(a<5 or a<6) # false or false
print(a>5 or a==6) # true or false


print(not a>5)

print(not(a>5 and a<6))
```

## 5.Membership Operators

```python
a = [1,2,3,4,5,6]
print(3 in a )
print(25 not in a)
print(3 not in a)

b ="Python"
print('p' in b)
print('P' in b)
```

## 6.Identity Operator

```python
# integer value

a = 12
print(id(a))

b = 12
print(id(b))

print(a is b)
```

```python
a = 257
print(id(a))

b = 257
print(id(b))

print(a is b)
```

```python
# float datatype

a = 23.5
b = 23.5
print(id(a))
print(id(b))
print(a is b)
print(a is not b)
```

```python
#Strings----> it will give same id for strings (length<=20)
a = 'hello'
print(len(a))
b = "hello"
print(len(b))
print(id(a))
print(id(b))
print(a is b)
```

```python
c = "this is a long string"
print(len(c))
d = "this is a long string"
print(len(d))
print(id(c))
print(id(d))
print(c is d)
```
