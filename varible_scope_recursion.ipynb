## local scope


```python
def greet():
    message="hello"
    print(message)
greet()
#message(hello)---->error,message is created inside a function,so exits while the function runs
```

```python
def show_name():
    name="jhon"
    print("name is:",name)
show_name()
```

```python
def caluclate_area(radius):
    pi=3.14
    area=pi*radius**2
    return area
result=caluclate_area(6)
result
```

## global varible

```python
app="flipkart"
def app_name():
    global app
    app="amazon"
    print(app)
app_name()
print(app)
   
```

```python
status="offline"
def go_online():
    global status
    status="online"
    print("status:",status)
go_online()
print("status2:",status)
```

## enclosing scope/non local varible

```python
def outer():
    x=10 #this varible is local to "outer()" non local to "inner()"
    def inner():
        print(x)
    inner()
    print(x)
outer()
```

## name spaces/LEGB RULE

```python
pi=3.1415 #global varible
def geomentry():
     #enclosing namespace
    pi=3.14 # enclosing varible
    def area(radius):
        #local name space
        shape='circle' #local varible
        result=pi*radius**2#take enclosong pi value
        print(f"pi value:{pi}")#enclosing pi
        print(f"shape:{shape}")#local shape
        print(f"area of circle:{result}")#locaal result
    area(5)
geomentry()
print("global varible pi value pi value:",pi)
```

## recursion 

```python
# without recursuion
def factorial_iter(n):
    result=1
    for i in range(1,n+1):
        result*=i
    return result
factorial_iter(5)
```

```python
#with recursion
def factorial_recursive(n):
    if n==0:
        return 1
    else:
        return n*factorial_recursive(n-1)
factorial_recursive(5)
```

```python
def rev_str(s):
    return s[::-1]
print(rev_str("hello"))
```

```python
#with recursion 
def rev_string(s):
    if len(s)<=1:#empty srting or 1 charcter
        return s
    else:
        return s[-1]+rev_string(s[:-1])
print(rev_string("hello"))
```
