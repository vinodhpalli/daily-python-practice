```python
#*Using args
 #Write a function multiply(*nums) that multiplies all numbers passed to it. Call it with 2, 3, 4 and print the result.
def mul(*args):
    result=1
    for i in args:
        result*=i
    return result
mul(2,3,4)
```

```python
#**Using kwargs
 #Write a function student_info(**details) that prints each key-value pair passed (e.g., name="Anu", age=20).
def info(**kwargs):
    for key,value in kwargs.items():
        print(key,":",value)
info(name='anu',age=20,city="hyderabad",marks=99)


```

```python
#1️. Store and Reuse Output (Return Concept)
#Write a function that takes the price of an item and GST percentage, calculates the final amount, and returns it.
 #Store the returned value in a variable and print the bill amount.
def amount(price,gst):
    final_price=price+gst
    return final_price
total_amount=amount(2000,50)
print(total_amount)
```
