# Higher-Order Functions

```python
#Given a list of numbers, write a program to double each number in the list.
a=[10,20,30]
result=list(map(lambda i:i*2,a))
print(result)

```

```python
#Given a list of numbers, write a program to keep only the numbers greater than 10.
a=[1,2,3,33,4,455,66,77,88,10]
result=list(filter(lambda i:i>10,a))
print(result)
```

```python
#Given a list of numbers, write a program to calculate the product of all numbers in the list.
from functools import reduce
a=[1,2,3,4,5,6,7,8]
result=reduce(lambda x,y:x*y,a)
print(result)
```

# Higher-Order Functions 

```python
#Given a list of numbers, write a program to triple each number using a single expression.

a=[10,20,30]
result=list(map(lambda i:i*3,a))
print(result)
```

```python
#From a list of numbers, write a program to keep only the odd numbers using a single expression
a=[1,2,3,4,5,6,7,8,9]
result=list(filter(lambda x:x%2==1,a))
print(result)
```

```python
#From a list of numbers, write a program to calculate the sum of all squares of the numbers.
from functools import reduce
a=[1,2,3,4,5,6,7,8,9]
result=list(map(lambda x:x**2,a))
sum_a=reduce(lambda x,y:x+y,result)

print(sum_a)

```

```python
#Given a list of strings, write a program to find the length of each string in a single expression.
a=["vinodh","durga","chandu"]
result=list(map(lambda x:len(x),a))
print(result)

```

```python
#From a list of numbers, write a program to find the largest number using a single expression.
a=[1,2,3,4,5,6,7,8,9]
result=reduce(lambda x,y:x if x>y else y,a)
print(result)
```
