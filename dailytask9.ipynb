```python
#6. Simulate ATM PIN Retry (3 Attempts Max)
 #Ask the user to enter their PIN. Allow only 3 incorrect attempts before locking the system.
  #Input: 123 → 000 → 456
  #Output: Account locked!
a=123
attempt=1
while True:
    b=int(input("enter a number:"))
    if b==a:
        print("pin is correct")
        break
    if attempt==3:
        print("account is locked")
        break
    attempt+=1
        
```

```python
#8. Sum All Odd Numbers Till n
 #Ask the user for a number n. Use a while loop to sum all odd numbers between 1 and n.
  #Input: 10
  #Output: 25
i=1
a=int(input("enter the number:"))
sum_a=0
while i<a:
    if i%2==1:
        sum_a+=i
    i+=1
print(sum_a)
```

```python
#12. Keep Accepting Numbers, Count How Many Were Even
 #Stop the loop when the user enters -1.
  #Input: 2 → 3 → 4 → 7 → -1
  #Output: Even numbers: 2
count=0
while True:
    a=int(input("enter the numbers:"))
    if a<0:
          break
    if a%2==0: 
        count+=1
print("even numbers:",count)

```
