```python
#1.  Scenario: Email Format Validator
#Create a program to check if the entered email is valid:
#Must contain @
#Must end with .com
#No starting blank spaces

email = input("Enter the email:")
if " " in email:
    print("Not valid email")
elif email.endswith(".com"):
    if '@' in email:
        print("Vald email")
    else:
        print("Not valid email")
else:
    print("Not valid email")
```

```python
#2.  Scenario: Username Format Checker
#Create a program to validate username format:
#Must start with a lowercase letter
#Can contain letters or  numbers or both 
#No symbols or spaces
name = input("Enter the username:")
if name[0].lower() and name.isalnum():
    print("Username accecpted")
else:
    print("Invalid username")
```

```python
#3.  Scenario: File Format Identifier
#Write a program to classify file type based on its extension:
#.pdf ➞ PDF File
#.docx ➞ Word File
#.xlsx ➞ Excel File
#Any other → Unknown Format
i_d = input("Enter the file format").lower()
if i_d == ".pdf":
    print("PDF file")
elif i_d == ".docx":
    print("Word file")
elif i_d == ".xlsx":
    print("Excel file")
else:
    print("Unkown format")
```

```python
#4.  Scenario: Blank Input Detector
#Create a program to detect if the input is blank or whitespace:
#Only whitespace → Blank Input
#Else → Input Received
a = input("Give an input:")
if " " in a :
    print("Blank input")
elif "" in a:
    print("Input received")
else:
    print("Input recived")
```

```python
#5.  Scenario: Proper Noun Validator
#Create a program to check if a word is a properly formatted proper noun:
#Starts with capital letter
#Remaining letters are lowercase
i = input("Enter the word:")
if i.title() and i.islower():
    print("Proper noun")
else:
    print("Incorrect capitilization")
```

```python
#6.  Scenario: Mobile Number Validator
#Write a program to validate a mobile number:
#Exactly 10 digits
#Only numeric characters allowed
num  = input("enter the number:")
if len(num) == 10 and num.isdigit():
    print("Valid mobile number")
    
else:
    print("Invalid number")
```

```python
#Scenario: Password Strength Checker
#Create a password checker with these rules:
#Minimum 8 characters
#Must contain both letters and numbers
password = input("Enter the password")
if  len(password) >= 8 and password.isalnum() and not password.isalpha() and not password.isdigit():
    print("Strong password")
else:
    print("Weak password")
```

```python
#Scenario: Sentence Format Checker
#Check if the sentence structure is proper:
#Starts with a capital letter
#Ends with a period .
a = input("Enter the sentence:")
if a.istitle() and a.endswith("."):
    print("Well-structured")

else:
    print("Needs correction")
```

```python
#10.  Scenario: Character Category Checker
#Write a program to categorize a single character:
#Alphabet
#Digit
#Space
#Special Character
a = input("Enter the character").lower()
if a.isalpha():
    print("Alphabet")
elif a.isdigit():
    print("Digit")
elif a == " ":
    print("Space")
else:
    print("Special Character")

```

```python
#11.  Scenario: Country Code Identifier
#Identify the country based on phone number prefix:
#+91 ➞ India
#+44 ➞ UK
#Others ➞ Other
num = input("Enter the number:")
if num.startswith("+91"):
    print("India")
elif num.startswith("+44"):
    print("UK")
else:
    print(" Other")


```

```python
#Scenario: Priority Message Classifier
#Classify messages by urgency based on keywords:
#Contains urgent, immediately ➞ High Priority
#Contains report, update ➞ Informational
#Else ➞ Normal
a = input("Enter the message:")
if "immediately" and "urgent" in a:
    print("High Priority")
elif "report" and "update" in a:
    print("Informational")
else:
    print(" Normal")
```

```python
#Scenario: Coupon Code Validator
#Validate a coupon code using the following rules:
#Must start with "SAVE"
#Only alphanumeric characters allowed
code = input("Enter the code")
if code.startswith("SAVE") and code.isalnum():
    print("Coupon Applied")
else:
    print("Invalid Coupon")
```

```python
#Scenario: Word Length Categorizer
#Categorize a word based on its character count:
#Length ≤ 3 → Short
#Length 4–6 → Medium
#Length > 6 → Long
word = input("Enter the word:")
if len(word) <= 3:
    print("Short")
elif len(word) >=4 and len(word) <= 6:
    print("Medium")
else:
    print("Long")
```
