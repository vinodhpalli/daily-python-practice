## list comprehensions

```python
#without list compershsion 
#generate a list of squres
squares=[]
for i in range(1,6):
    squares.append(i**2)
print(squares)
```

```python
#with list compershion
squares=[i**2 for i in range(1,15)]
print(squares)
```

```python
#generate even numbers
even=[]
for i in range(1,20):
    if i%2==0:
        even.append(i)

print(even)
```

```python
even=[i for i in range(1,23) if i%2==0]
print(even)
```

```python
#converet all strings to upperclss
name=['apple','bananan','orange']
new=[i.upper() for i in name]
print(new)
```

```python
pairs=[(i,i**2) for i in range(1,8)]
print(pairs)
```

```python
#nested list cophersion
matrix=[[i*j for j in range(1,4)] for i in range(1,4)]
print(matrix)
```
